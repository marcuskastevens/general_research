# Get Normalized PE

```python
pe_ratio = key_metrics_annually.loc['peRatio'].loc[:'1995']
pe_ratio.loc['2022'] = 15.25
pe_ratio = pe_ratio.loc[['2022', '2021', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013',
       '2012', '2011', '2010', '2009', '2008', '2007', '2006', '2005', '2004',
       '2003', '2002', '2001', '2000', '1999', '1998', '1997', '1996', '1995']]

normalized_pe_ratio = (pe_ratio - pe_ratio.mean()) / pe_ratio.std()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 9))
plt.plot(normalized_pe_ratio, label='HIFS P/E Ratio')
plt.title('Z-Score Normalized P/E Ratio')
plt.axhline(1.28, linestyle='--', color='r', label='90% CI')
plt.axhline(-1.28, linestyle='--', color='g', label='10% CI')
plt.legend()
```

# Get Normalized ROE

```python
roe = financial_ratios_annually.loc['returnOnEquity'].loc[:'1995']
roe.loc['2022'] = .1148
roe = roe.loc[['2022', '2021', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013',
       '2012', '2011', '2010', '2009', '2008', '2007', '2006', '2005', '2004',
       '2003', '2002', '2001', '2000', '1999', '1998', '1997', '1996', '1995']]

normalized_roe = (roe - roe.mean()) / roe.std()
```
